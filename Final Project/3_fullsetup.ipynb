```python
import os
import cv2
import numpy as np

train_dir = "/home/nick/Documents/Datasets/ASL/asl_alphabet_train"
test_dir = "/home/nick/Documents/Datasets/ASL/test_set"

def get_train_data(data_dir) :
    images = []
    labels = []
    
    dir_list = os.listdir(data_dir)
    dir_list.sort()
    print(dir_list)
    label_eq = dir_list
    for i in range(len(dir_list)):
        count = 0
        print("Obtaining images of", dir_list[i], " size: ", len(os.listdir(data_dir + "/" + dir_list[i])))
        for image in os.listdir(data_dir + "/" + dir_list[i]):
            if (count+1) % 3 != 0: # Dataset is too big for local memory, taking 2 of every 3 images.
                img = cv2.imread(data_dir + '/' + dir_list[i] + '/' + image)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (125, 125), interpolation = cv2.INTER_AREA)
                img1 = np.copy(img)
                img2 = np.copy(img)
                img3 = np.copy(img)

                img1[:][:][0] = 0.5*img1[:][:][0]
                img1[:][:][1] = 0.5*img1[:][:][1]

                img2[:][:][1] = 0.5*img2[:][:][1]
                img2[:][:][2] = 0.5*img2[:][:][2]

                img3[:][:][0] = 0.5*img3[:][:][0]
                img3[:][:][2] = 0.5*img3[:][:][2]
        
                images.append(img)
                labels.append(i)
                images.append(img1)
                labels.append(i)
                images.append(img2)
                labels.append(i)
                images.append(img3)
                labels.append(i)
            count += 1
    
    return images, labels, label_eq

def get_test_data(data_dir) :
    images = []
    labels = []
    
    dir_list = os.listdir(data_dir)
    dir_list.sort()
    print(dir_list)
    label_eq = dir_list
    for i in range(len(dir_list)):
        print("Obtaining images of", dir_list[i], " size: ", len(os.listdir(data_dir + "/" + dir_list[i])))
        for image in os.listdir(data_dir + "/" + dir_list[i]):
            img = cv2.imread(data_dir + '/' + dir_list[i] + '/' + image)
            img = cv2.resize(img, (125, 125), interpolation = cv2.INTER_AREA)
            images.append(img)
            labels.append(i)
    
    return images, labels, label_eq
        
XT, YT, label_eq_train = get_train_data(train_dir)
XTE, YTE, label_eq_test = get_test_data(test_dir)
```

```python
for i in range(len(XT)):
    XT[i] = XT[i].T
XT = np.array(XT)
YT = np.array(YT)

for i in range(len(XTE)):
    XTE[i] = XTE[i].T
XTE = np.array(XTE)
YTE = np.array(YTE)
```

```python
import h5py
file = '3_fullimg.hdf5'
with h5py.File(file, 'w') as hf:
    hf.create_dataset('x_train', data = np.asarray(XT))
    hf.create_dataset('x_test', data = np.asarray(XTE))
    hf.create_dataset('y_train', data = np.asarray(YT))
    hf.create_dataset('y_test', data = np.asarray(YTE))
```
